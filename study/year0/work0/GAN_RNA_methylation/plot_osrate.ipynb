```python
import os
import pandas as pd
import numpy as np
```

```python
RESULT_DIR = 'result'
data = pd.DataFrame(columns=['model', 'Accuracy', 'os_rate'])
for os_rate in np.arange(0., 1., 0.10):
    c_dir = os.path.join(RESULT_DIR, f'GAN_test_mcc_{os_rate:.1f}.csv')
    if os.path.exists(c_dir):
        # print(c_dir)
        result = pd.read_csv(c_dir, sep='\t')
        result = result[['model', 'Accuracy']]
        result["os_rate"] = os_rate
        result = result.iloc[:5,]
        data = data.append(result)
data
```

```python
from plotnine import *
```

```python
plot = ggplot(data, aes(x='os_rate', y='Accuracy', shape='model',color='model')) + xlim(0.1, 0.9) + ylim(0.85, 0.98)
plot += geom_line()
plot += geom_point(size=2)
plot += theme_light()
plot += scale_x_continuous(breaks=np.arange(0,1,0.1))
plot += xlab('')
ggsave(plot,"plot_osrate.svg")
plot
```
